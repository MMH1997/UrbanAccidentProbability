### Create lists to add the metrics

```python
IAN = []
Acu = []
Bal = []
```

### Perform by hour the same than in the main experiment:
#### 1. Upload the dataframe (df) and remove any missing values and the first column (if necessary).
#### 2. Define predictor variables (excluding the hour and related variables) and apply an oversampling technique.
#### 3. Apply the CatBoost (CB) model and obtain predictions.
#### 4. Calculate the metrics. 

```python
for i in range(1,25):
    df = pd.read_csv('DF_FINAL.csv')
    df = df.drop(columns=['Unnamed: 0'])
    df = df.dropna()
    df = df[df['Hora'] == i]
    df

    X = df[['Mes', 'Dia', 'Dia_S', 'Trafico', 'Roads', 'tmed', 'prec', 'velmedia']]
    y = df['Accidentes']

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33,random_state = 44)
    smote = SMOTE(sampling_strategy='not minority', k_neighbors = 9, random_state=42)
    X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

    catboost_model = CatBoostClassifier()
    catboost_model.fit(X_train_resampled, y_train_resampled)

    pr = [lista[1] for lista in catboost_model.predict_proba(X_test)]
    # probabilidades = svm_classifier.predict_proba(X_test)[:, 1]
    data = {'Columna1': y_test.values,
            'Columna2': pr}
    df1 = pd.DataFrame(data)

    probi = []
    for i in range(len(df1)):
        if df1.iloc[i,1] > 0.1:
            probi.append(1)
        else:
            probi.append(0)

    IAN.append(df1.loc[df1['Columna1'] != 0, 'Columna2'].mean()/df1.loc[df1['Columna1'] == 0, 'Columna2'].mean())
    Acu.append(accuracy_score(y_test, probi))
    Bal.append(balanced_accuracy_score(y_test, probi))
```

### Print metric values.

```python
print(np.mean(IAN), np.mean(Acu), np.mean(Bal))
```
