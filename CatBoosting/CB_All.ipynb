### Upload dataset and, if necessary, remove the first column and NA values.

```python
df = pd.read_csv('DF_FINAL.csv')
df = df.drop(columns=['Unnamed: 0'])
df = df.dropna()
df
```

### Define the predictor variables and the target variable. 

```python
X = df[['Mes', 'Dia', 'Dia_S', 'Hora', 'seno_horas', 'coseno_horas', 'Trafico', 'Roads', 'tmed', 'prec', 'velmedia']]
y = df['Accidentes']
```

### Define the training and testing sets and apply an oversampling technique.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33,random_state = 44)
smote = SMOTE(sampling_strategy='not minority', k_neighbors = 9, random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

### Apply the CatBoost Classifier model to the resampled training set and predict probabilities in the testing set.

```python
catboost_model = CatBoostClassifier()
catboost_model.fit(X_train_resampled, y_train_resampled)

# Predecir probabilidades en el conjunto de prueba
catboost_probs = catboost_model.predict_proba(X_test)
```

### Code to obtain the I_{A/C} metric.

```python
pr = [lista[1] for lista in catboost_model.predict_proba(X_test)]
probabilidades = catboost_model.predict_proba(X_test)
data = {'Columna1': y_test.values,
        'Columna2': pr}
df1 = pd.DataFrame(data)

probi = []
for i in range(len(df1)):
    if df1.iloc[i,1] > 0.1:
        probi.append(1)
    else:
        probi.append(0)


print(df1.loc[df1['Columna1'] != 0, 'Columna2'].mean())
print(df1.loc[df1['Columna1'] == 0, 'Columna2'].mean())
```

### I_{A/C} metric

```python
df1.loc[df1['Columna1'] != 0, 'Columna2'].mean()/df1.loc[df1['Columna1'] == 0, 'Columna2'].mean()
```

### Accuracy and Balanced accuracy metrics

```python
print(accuracy_score(y_test, probi))
print(balanced_accuracy_score(y_test, probi))
```
